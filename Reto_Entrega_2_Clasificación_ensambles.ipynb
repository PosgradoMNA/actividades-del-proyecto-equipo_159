```python
#Borrar columnas vacías
datos.drop("TOX_D_48_FON_UT", inplace=True, axis=1)
datos.drop("CALIDAD_TOX_D_48_FON", inplace=True, axis=1)
datos.drop("TOX_FIS_FON_15_UT", inplace=True, axis=1)
datos.drop("CALIDAD_TOX_FIS_FON_15", inplace=True, axis=1)

#Borrar columnas con pocos records (487)
datos.drop("OD_PORC_MED", inplace=True, axis=1)
datos.drop("CALIDAD_OD_PORC_MED", inplace=True, axis=1)

#Borrar columna con información poco relevante
datos.drop("PERIODO", inplace=True, axis=1)
datos.drop("CONTAMINANTES", inplace=True, axis=1)
```

```python
#CUENCA
df = pd.DataFrame(datos,columns=["CUENCA", "ESTADO", "MUNICIPIO"])
nan_values = df[df["CUENCA"].isna()]

#Encontrar records en los datos con el mismo Municipio y Estado
michoacan_records = df.loc[ (df['MUNICIPIO'] == 'LAZARO CARDENAS') & (df['ESTADO'] == 'MICHOACAN DE OCAMPO')]

#Encontrar los valores unicos para CUENCA en los records anteriormente encontrados y mostrar el valor máximo
max_cuenca = michoacan_records["CUENCA"].value_counts().idxmax()

#Reemplazar el valor vacio de CUENCA con el máximo para este dato
datos["CUENCA"].replace(np.nan, max_cuenca, inplace=True)
```

```python
#borrar todos los CUERPO DE AGUA debido a que todos los records para esa Cuenca en ese Municipio dentro de 
#ese estado están vacios
datos.dropna(subset=["CUERPO DE AGUA"], axis=0, inplace=True)
```

```python
#borrar todos los LONGITUD y LATITUD   debido a que todos los records para esa 
#Cuenca en ese Municipio dentro de ese estado están vacios
datos.dropna(subset=["LONGITUD"], axis=0, inplace=True)  
datos.dropna(subset=["LATITUD"], axis=0, inplace=True)  
```

```python
#TIPO y SUBTIPO
df = pd.DataFrame(datos,columns=["TIPO", "SUBTIPO"])
nan_values = df[df["SUBTIPO"].isna()]
print (nan_values)
```

```python
datos = datos.replace('<1','1')
datos = datos.replace('<2','2')
datos = datos.replace('<3','3')
datos = datos.replace('<10','10')
```

```python
datos['DQO_mg/L'] = datos['DQO_mg/L'].astype('float64')
datos['SST_mg/L'] = datos['SST_mg/L'].astype('float64')
datos['COLI_FEC_NMP_100mL'] = datos['COLI_FEC_NMP_100mL'].astype('float64')
datos['E_COLI_NMP_100mL'] = datos['E_COLI_NMP_100mL'].astype('float64')
datos['OD_PORC'] = datos['OD_PORC'].astype('float64')
datos['OD_PORC_SUP'] = datos['OD_PORC_SUP'].astype('float64')
datos['OD_PORC_FON'] = datos['OD_PORC_FON'].astype('float64')

datos['DBO_mg/L'] = datos['DBO_mg/L'].astype('float64')
datos['ENTEROC_NMP_100mL'] = datos['ENTEROC_NMP_100mL'].astype('float64')
datos['TOX_D_48_UT'] = datos['TOX_D_48_UT'].astype('float64')
datos['TOX_V_15_UT'] = datos['TOX_V_15_UT'].astype('float64')
datos['TOX_D_48_SUP_UT'] = datos['TOX_D_48_SUP_UT'].astype('float64')
datos['TOX_FIS_SUP_15_UT'] = datos['TOX_FIS_SUP_15_UT'].astype('float64')
```

```python
#Consideramos que para llenar los datos nulos de las variables numericas se debe realizar basado en el semaforo, 
#por lo cual realizaremos 3 clasificaciones

semaforo_verde = pd.DataFrame(datos[datos['SEMAFORO'] == 'Verde'])
semaforo_amarillo = pd.DataFrame(datos[datos['SEMAFORO'] == 'Amarillo'])
semaforo_rojo = pd.DataFrame(datos[datos['SEMAFORO'] == 'Rojo'])
```

```python
Numericas = ['DBO_mg/L','DQO_mg/L','SST_mg/L','COLI_FEC_NMP_100mL','E_COLI_NMP_100mL','ENTEROC_NMP_100mL','OD_PORC_SUP', 'OD_PORC',
             'OD_PORC_FON','TOX_D_48_UT','TOX_V_15_UT','TOX_D_48_SUP_UT','TOX_FIS_SUP_15_UT']
```

```python
#Calcular media de variables Numericas para reemplazar valores vacíos
#for para calcular y reemplazar valores

mean_verdes = list()
mean_rojos = list()
mean_amarillos = list()

for n in Numericas:
  mean1 = semaforo_verde[n].mean(skipna=True)
  semaforo_verde[n].replace(np.nan, mean1, inplace=True)
  mean_verdes.append(mean1)

for n in Numericas:
  mean1 = semaforo_amarillo[n].mean(skipna=True)
  semaforo_amarillo[n].replace(np.nan, mean1, inplace=True)
  mean_amarillos.append(mean1)

for n in Numericas:
  mean1 = semaforo_rojo[n].mean(skipna=True)
  semaforo_rojo[n].replace(np.nan, mean1, inplace=True)
  mean_rojos.append(mean1)
```

```python
Categoricas = ["CALIDAD_DBO", "CALIDAD_DQO", "CALIDAD_SST", "CALIDAD_COLI_FEC", "CALIDAD_E_COLI", "CALIDAD_ENTEROC", "CALIDAD_OD_PORC_SUP",  "CALIDAD_OD_PORC",
               "CALIDAD_OD_PORC_FON", "CALIDAD_TOX_D_48", "CALIDAD_TOX_V_15", "CALIDAD TOX_D_48_SUP",
               "CALIDAD_TOX_FIS_SUP_15"]
```

```python
#Calcular moda de variables Numericas para reemplazar valores vacíos
#for para calcular y reemplazar valores
mode_verdes = list()
mode_rojos = list()
mode_amarillos = list()

for n in Categoricas:
  mode1 = semaforo_verde[n].mode()
  semaforo_verde[n].replace(np.nan, mode1[0], inplace=True)
  mode_verdes.append(mode1)

for n in Categoricas:
  mode1 = semaforo_amarillo[n].mode()
  semaforo_amarillo[n].replace(np.nan, mode1[0], inplace=True)
  mode_amarillos.append(mode1)

for n in Categoricas:
  mode1 = semaforo_rojo[n].mode()
  semaforo_rojo[n].replace(np.nan, mode1[0], inplace=True)
  mode_rojos.append(mode1)

```

```python
#Inner Join Code
data_clean = pd.concat([semaforo_verde, semaforo_amarillo, semaforo_rojo],join='inner')
print(data_clean.head())
datos = data_clean
```

# Seleccionamos las variables independientes X y dependiente Y (semáforo)

```python
X = datos[['CLAVE', 'SITIO', 'ORGANISMO_DE_CUENCA', 'ESTADO', 'MUNICIPIO',
           'CUENCA', 'CUERPO DE AGUA', 'TIPO', 'SUBTIPO', 'LONGITUD',
           'LATITUD', 'DBO_mg/L', 'CALIDAD_DBO', 'DQO_mg/L', 'CALIDAD_DQO',
           'SST_mg/L', 'CALIDAD_SST', 'COLI_FEC_NMP_100mL',
           'CALIDAD_COLI_FEC', 'E_COLI_NMP_100mL', 'CALIDAD_E_COLI',
           'ENTEROC_NMP_100mL', 'CALIDAD_ENTEROC', 'OD_PORC',
           'CALIDAD_OD_PORC', 'OD_PORC_SUP', 'CALIDAD_OD_PORC_SUP',
           'OD_PORC_FON', 'CALIDAD_OD_PORC_FON', 'TOX_D_48_UT',
           'CALIDAD_TOX_D_48', 'TOX_V_15_UT', 'CALIDAD_TOX_V_15',
           'TOX_D_48_SUP_UT', 'CALIDAD TOX_D_48_SUP', 'TOX_FIS_SUP_15_UT',
           'CALIDAD_TOX_FIS_SUP_15','CUMPLE_CON_DBO','CUMPLE_CON_DQO', 
           'CUMPLE_CON_SST', 'CUMPLE_CON_CF','CUMPLE_CON_E_COLI', 
           'CUMPLE_CON_ENTEROC', 'CUMPLE_CON_OD', 'CUMPLE_CON_TOX', 'GRUPO']]

y = datos[['SEMAFORO']]
```

```python
#Usamos label encoder, para convertir las clases del semáforo de categóricas a numéricas.
from sklearn import preprocessing
l_encoder = preprocessing.LabelEncoder()

y['SEMAFORO']=l_encoder.fit_transform(y['SEMAFORO'])
```

```python
#Transformamos las variables de semaforo a 0, 1, 2.
l_encoder.inverse_transform([0,1,2])
```

```python
#Convertimos todas las variables categóricas a numéricas mediante un fit
for c in X.columns:
  X[c]=l_encoder.fit_transform(X[c])
```

```python
#Generamos un bosque aleatorio para revisar la importancia de las columnas para el algortimo

from sklearn.ensemble import RandomForestClassifier

RFC = RandomForestClassifier(n_estimators=100, random_state=42)
RFC.fit(X, y)

score_list=[]
name_list=[]

for score, name in zip(RFC.feature_importances_, X.columns):
  score_list.append(score)
  name_list.append(name)

scores = pd.DataFrame(name_list,score_list)
scores.sort_index(ascending=False)
```
